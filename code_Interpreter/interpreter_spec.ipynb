# Commands for getting Interpreter Specification

```python
from bedrock_agentcore.tools.code_interpreter_client import CodeInterpreter

code_client = CodeInterpreter('us-east-1')
code_interpreter_id = "aws.codeinterpreter.v1"
session_id = code_client.start(identifier=code_interpreter_id)

def print_respoonse(response):
    for event in response["stream"]:
        result = event["result"]
        for content in result["content"]:
            print(content["text"])
```

### Basic Hardware information

```python
response = code_client.invoke("executeCommand", {
    "command": "uname -a",
})
print_respoonse(response)
```

### Possible commands

```python
response = code_client.invoke("executeCommand", {
    "command": "compgen -c",
})
print_respoonse(response)
```

### Python Environment

```python
response = code_client.invoke("executeCommand", {
    "command": "python --version && pip --version",
})
print_respoonse(response)
```

```python
response = code_client.invoke("executeCommand", {
    "command": "pip list",
})
print_respoonse(response)
```

### Clean up

```python
code_client.stop()
```
